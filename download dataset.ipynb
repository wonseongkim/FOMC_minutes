```python
import requests
import pandas as pd

def download_dataset(url, file_name):
    # Make a GET request to the specified URL
    response = requests.get(url)
    # Check if the request was successful (HTTP status code 200)
    if response.status_code == 200:
        # Open the file with the specified file name in write-binary mode
        with open(file_name, 'wb') as f:
            # Write the content of the response to the file
            f.write(response.content)
        # Correctly use the file_name variable in the print statement
        print(f"Download complete: {file_name}")
    else:
        # Inform the user if the download failed
        print("Failed to download file")

# Specify the URL of the dataset
url = "https://raw.githubusercontent.com/wonseongkim/FOMC_minutes/main/fomc_minutes-label.csv"
# Specify the file name for the downloaded dataset
file_name = "fomc_minutes-label.csv"

# Call the function to download the dataset
download_dataset(url, file_name)

# Load the CSV file into a DataFrame
df = pd.read_csv(file_name)

df
```
